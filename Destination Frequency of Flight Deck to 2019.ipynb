```python
#first work with FlightDeck
FD = f[f.CrewType=='FlightDeck']
```

```python
#all unique positions of Flight Deck: CP = Captain, FO = First Officier;
FD.Position.unique()
```

```python
#Bases where located Flight Deck
FD.Base.unique()
```

```python
# Select only flight deck which based in ALA
FDALA=FD[FD['Base'].str.strip()=='ALA']
```

```python
FDALA.head()
```

```python
#Decided do not delete columns, just selected columns which I need to analyze
FDALA = FDALA[['CrewMemberID', 'Base','Position', 'Gender', 'CrewType',
       'FlightNo', 'FlightDate', 'Cancelled', 'FlightPostfix', 'DepartureAirport', 'ArrivalAirport', 
   'TailNumber', 'Fleet', 'AcType']]
```

```python
# All Pilots which qualified to flight on Airbus Aircrafts
Airbus = FDALA[(FDALA['Fleet'] == '320' ) 
                  | (FDALA['Fleet']== '321' ) 
                  | (FDALA['Fleet'].str.strip()== '321L' )
                  | (FDALA['Fleet']== '319' )] 
```

```python
Airbus.ArrivalAirport.unique()
```

```python
Airports = pd.read_csv("Airports.csv")
```

```python
# All Kazakhstan's Airport
Domestic = Airports[Airports.COUNTRY=='UA']
```

```python
Domestic.head()
```

```python
# Choose only International flights
Airbus = Airbus[~ Airbus.ArrivalAirport.isin(Domestic['AIRPORT'])]
```

```python
Airbus.head()
```

```python
FrequencyofAirbus = Airbus.pivot_table(index=['CrewMemberID'], columns=['ArrivalAirport'], values='Base', aggfunc='count')
```

```python
Airbus.ArrivalAirport.unique()
```

```python
#change all columns to upper case
FrequencyofAirbus.rename(columns = lambda x:x.upper(), inplace = True) 
```

```python
#all NaN change to -> 0 
FrequencyofAirbus=FrequencyofAirbus.fillna(0)
```

```python
#convert float to int
FrequencyofAirbus=FrequencyofAirbus.astype(int)
```

```python
#The Result for Airbus
FrequencyofAirbus
```

```python
#here the flight decks which had several flights for current destinations, send to Planning team, to do not plan this destinations to Crew
FrequencyofAirbus1=FrequencyofAirbus[FrequencyofAirbus>5]
```
